```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons#Set of points in 2d making
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
X,_=make_moons(n_samples=200,noise = 0.1)
#Standardize the DataSet
scaler =StandardScaler()
X_scaled=scaler.fit_transform(X)

```

```python
#Visualize the Clustering Result
plt.scatter(X[:, 0], X[:, 1])
```

```python
#initialize The DBSCAN ALGORITHM PARAMETERS I:E Minimum no of data points and radius/Epsilon
Clus=DBSCAN(eps=0.4,min_samples=5)
#perform Clustering
labels = Clus.fit_predict(X_scaled)
#Plot the Results
```

```python
#Plot the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Result for DBSCAN Clustering")
plt.show()
```
