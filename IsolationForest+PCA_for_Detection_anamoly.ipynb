```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
np.random.seed(42)
normal_data=np.random.normal(loc=0,scale=1,size=(100,2))
anomalous_data=np.random.normal(loc=4,scale=1,size=(10,2))
all_data=np.vstack((normal_data,anomalous_data))

```

```python
#Train the isolation forest model
clf=IsolationForest(contamination=0.05,random_state=42)
clf.fit(all_data)
```

```python
#Predict anomalies
predictions=clf.predict(all_data)
anomaly_indices=np.where(predictions==-1)[0]
```

```python
#Plot the Data and Anamolies
plt.scatter(all_data[:,0],all_data[:,1],c='blue',label='Normal Data')
plt.scatter(all_data[anomaly_indices,0],all_data[anomaly_indices,1],c='red',label='Anamolies')
plt.legend()
plt.title("Anamoly Detection Using Isolation Forest")
plt.xlabel('feature -1')
plt.ylabel('feature -2')
plt.show()
```
