```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import  dendrogram, linkage
```

```python
data, labels = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=0)
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title('Original Data')
plt.show()

```

```python
#perform Agglomerative Clustering
a1=AgglomerativeClustering(n_clusters=3)
labels=a1.fit_predict(data)

#Visualize the Clustering Result
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title("Agglomerative Clustering")
plt.show()
```

```python
#Create A dendogram
linked = linkage(data,'ward')#Ward is a minimum variance crieterion
dendrogram(linked)
plt.title("Clustering DendoGram")
plt.show()
```
