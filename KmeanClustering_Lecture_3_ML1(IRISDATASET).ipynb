```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
```

```python
# Apply KMeans clustering
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
```

```python
# Plotting Sepal Length vs Sepal Width
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('KMeans Clustering on Iris Dataset')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features

# Apply KMeans clustering
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)
labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_  # Retrieve cluster centers

# Visualize the clusters
plt.figure(figsize=(10, 6))

# Plotting Sepal Length vs Sepal Width with cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, label='Data Points')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('KMeans Clustering on Iris Dataset')
plt.legend()
plt.show()
```
